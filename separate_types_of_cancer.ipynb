```python
import pandas as pd
import os
```

```python
df = pd.read_csv("SAMPLE_INFO.txt", sep="\t")
print(df.head())
```

```python
output_dir = "cancer_samples"
os.makedirs(output_dir, exist_ok=True)
```

```python
types_of_cancer = df["sj_diseases"].dropna().unique()
print(types_of_cancer)
```

```python
for sample in types_of_cancer:
    df_tipo = df[df["sj_diseases"] == sample]
    output_path = os.path.join(output_dir, f"{sample}_samples.tsv")
    df_tipo.to_csv(output_path, sep="\t", index=False)
```
